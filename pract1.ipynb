```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,400)
def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return np.tanh(x)

def softmax(x):
    e = np.exp(x - np.max(x))
    return e/e.sum(axis=0)

def relu(x):
    return np.maximum(0,x)
```

```python
plt.subplot(2,2,1)
plt.plot(x,sigmoid(x),label="sigmoid")
plt.plot(x,tanh(x),label="tanh")
plt.xlabel("x")
plt.ylabel("activation")
plt.grid(True)
plt.legend()
plt
```

```python
plt.subplot(2,2,1)
plt.plot(x,relu(x),label="relu")
plt.xlabel("x")
plt.ylabel("activation")
plt.grid(True)
```

```python
plt.subplot(2,2,1)
plt.plot(x,softmax(x),label="softmax")
plt.xlabel("x")
plt.ylabel("activation")
plt.grid(True)
```
